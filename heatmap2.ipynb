```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("C:/Users/Candy/Desktop/internship/archive/Future50.csv")
data.head()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 选择需要分析的变量
#variables = ['Rank','Restaurant','Location','Sales','YOY_Sales','Units','YOY_Units','Unit_Volume','Franchising']
variables = ['Rank','Sales','YOY_Sales','Units','YOY_Units','Unit_Volume']
# 计算变量之间的相关系数
correlation_matrix = data[variables].corr()

# 创建热力图
sns.heatmap(correlation_matrix, annot=True, cmap='YlGnBu')

# 设置图形属性
plt.title('Correlation Heatmap')
plt.xlabel('Variables')
plt.ylabel('Variables')

# 显示图形
plt.show()
```

```python
df = pd.DataFrame(data)

# 使用get_dummies函数对"Restaurant"和"Location"列进行独热编码
encoded_df = pd.get_dummies(df, columns=['Restaurant', 'Location'])

# 将"Franchising"列转换为数字形式
encoded_df['Franchising_Code'] = encoded_df['Franchising'].map({'Yes': 1, 'No': 0})

# 选择需要分析的变量
variables = ['Rank', 'Sales', 'YOY_Sales', 'Units', 'YOY_Units', 'Unit_Volume', 'Franchising_Code']

# 计算变量之间的相关系数
correlation_matrix = encoded_df[variables].corr()

# 创建热力图
sns.heatmap(correlation_matrix, annot=True, cmap='YlGnBu')

# 设置图形属性
plt.title('Correlation Heatmap')
plt.xlabel('Variables')
plt.ylabel('Variables')

# 显示图形
plt.show()
```
