```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 创建随机的二维数据
data = np.random.rand(10, 10)

# 使用Seaborn绘制热图
sns.heatmap(data, cmap='hot', annot=True, fmt=".2f")

# 添加标题和标签
plt.title("Heatmap Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# 显示热图
plt.show()
```
