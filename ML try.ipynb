```python
df.columns = df.columns.str.strip()
df['Income'] = pd.to_numeric(df['Income'], errors='coerce')
mean_income = df['Income'].mean()
df['Income'].fillna(mean_income, inplace=True)
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'],errors = 'coerce')
df['Customer_year'] = df['Dt_Customer'].dt.year
df['Age'] = 2025 - df['Year_Birth']
df['Children'] = df['Kidhome'] + df['Teenhome']
df['Total_spent'] = df['MntWines'] + df['MntFruits']+ df['MntMeatProducts']+ df['MntFishProducts']+ df['MntSweetProducts']+ df['MntGoldProds']
df['Total_purchases'] = df['NumDealsPurchases'] + df['NumWebPurchases'] + df['NumCatalogPurchases'] + df['NumStorePurchases']+df['NumWebVisitsMonth']
df['AcceptedAny'] =  df['AcceptedCmp1'] + df['AcceptedCmp2']+ df['AcceptedCmp3']+ df['AcceptedCmp4']+ df['AcceptedCmp5']+df['Response']
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# df.info()
```

```python
selected_features = ['Income', 'Children', 'Total_spent', 'Total_purchases', 'AcceptedAny', 'Age', 'NumStorePurchases']
df_selected = df[selected_features]

correlation_matrix_1 = df_selected.corr().fillna(0)

num_store_purchases_correlations = correlation_matrix_1['NumStorePurchases']

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix_1, annot=True, cmap='coolwarm',square=False)
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.title('NumStorePurchases Correlation')
plt.show()
```

```python
# ç­›é€‰å‡ºæ˜¾è‘—ç›¸å…³æ€§ï¼ˆç›¸å…³æ€§ > 0.5 æˆ–ç›¸å…³æ€§ < -0.5ï¼‰
significant_correlations = num_store_purchases_correlations[(num_store_purchases_correlations > 0.5) | (num_store_purchases_correlations < -0.5)]

print('Correlation Factorï¼š')
print(significant_correlations)
```

```python
df['Is_US'] = df['Country'] == 'US'
plt.figure(figsize=(8, 6))
sns.boxplot(x='Is_US', y='Total_purchases', data=df)
plt.xlabel('Is the country US')
plt.xticks(rotation=45)
plt.ylabel('Total purchases')
plt.title('Comparison of total purchases between US and non - US regions')
plt.xticks([0, 1], ['Non - US', 'US'])
plt.show()
```

```python
average_gold_spending = df['MntGoldProds'].mean()
df['Above_avg_gold_spending'] = df['MntGoldProds'] > average_gold_spending
plt.figure(figsize=(8, 6))
sns.boxplot(x='Above_avg_gold_spending', y='NumStorePurchases', data=df)

# Set the axis labels and title
plt.xlabel('Above average gold spending')
plt.xticks(rotation=45)
plt.ylabel('Number of in - store purchases')
plt.title('Comparison of in - store purchases between above and below average gold spending groups')
plt.xticks([0, 1], ['Below average', 'Above average'])
plt.show()
```

```python
df['Is_Married_PhD'] = (df['Marital_Status'] == 'Married') & (df['Education'] == 'PhD')

plt.figure(figsize=(8, 6))
sns.boxplot(x='Is_Married_PhD', y='MntFishProducts', data=df)

plt.xlabel('Is Married PhD')
plt.xticks(rotation=45)
plt.ylabel('Spending on fish products')
plt.title('Comparison of spending on fish products between married PhD and non - married PhD groups')
plt.xticks([0, 1], ['Non - married PhD', 'Married PhD'])

plt.show()
```

```python
selected_columns = ['Children', 'Age', 'MntWines', 'MntMeatProducts', 'MntFruits', 'MntGoldProds', 'MntSweetProducts', 'MntFishProducts']
selected_df = df[selected_columns]
correlation_matrix_2 = selected_df.corr()
plt.rcParams['figure.dpi'] = 300
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix_2, annot=True, cmap='coolwarm')
plt.title('Fish Correlation')
plt.show()
```

```python
mnt_fish_products_correlations = correlation_matrix_2['MntFishProducts']

significant_correlations = mnt_fish_products_correlations[(mnt_fish_products_correlations > 0.5) | (mnt_fish_products_correlations < -0.5)]

print('Correlation Factorï¼š')
print(significant_correlations)
```

```python
country_acceptedany_sum = df.groupby('Country')['AcceptedAny'].sum().reset_index()

sns.barplot(data=country_acceptedany_sum, x='Country', y='AcceptedAny')
plt.title('AcceptedAny Sum Comparison by Country')
plt.show()
```

```python
campaign_columns = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5']

campaign_counts = df[campaign_columns].sum()
simplified_labels = [col.replace('Accepted', '') for col in campaign_columns]
plt.bar(simplified_labels,campaign_counts)
plt.title('Success of Different Marketing Campaigns')
plt.xlabel('Marketing Campaign')
plt.ylabel('Number of Acceptances')
plt.show()
```

```python
product_columns = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
product_performance = df[product_columns].sum()

plt.bar(product_columns, product_performance)
plt.title('Performance of Different Products')
plt.xlabel('Product')
plt.ylabel('Total Consumption Amount')

plt.xticks(rotation=45)

plt.show()
```
