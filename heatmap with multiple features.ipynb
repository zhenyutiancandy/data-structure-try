```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 选择需要分析的变量
variables = ['sepal length(cm)', 'sepal width(cm)', 'petal length(cm)', 'petal width(cm)', 'class']

# 计算变量之间的相关系数
correlation_matrix = pd_iris[variables].corr()

# 创建热力图
sns.heatmap(correlation_matrix, annot=True, cmap='YlGnBu')

# 设置图形属性
plt.title('Correlation Heatmap')
plt.xlabel('Variables')
plt.ylabel('Variables')

# 显示图形
plt.show()
```
